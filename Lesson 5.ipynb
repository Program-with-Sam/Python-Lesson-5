```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [random.randint(0,100) for i in range(250)]
data = np.asarray(data)
```

```python
# Plots using matplotlib
plt.subplot(321), plt.plot(data)
plt.subplot(322), plt.hist(data)
plt.subplot(323), plt.bar(data, 30)
plt.subplot(324), plt.pie(data)
plt.subplot(325), plt.scatter(data, y=[x for x in range(1,251)])
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Sam/Downloads/wine-quality-red.csv', index_col=0, header=0)
df.head()
```

```python
# Plots using seaborn (and matplotlib for subplotting)
plt.subplot(321), sns.swarmplot(x='alcohol', y='pH', data=df)
plt.subplot(322), sns.violinplot(x='alcohol', y='pH', data=df)
plt.subplot(323), sns.boxplot(data=df)
plt.subplot(324), sns.lmplot(x='volatile_acidity', y='citric_acid', data=df, fit_reg=False)
```
